Aps rodar o cdigo em anexo, faa um crculo branco no grfico do contedo em anexo, indicando os pontos $(x1, x2)$ tais que $x1^2$ + $x2^2$ $\sim$ 10. Ou seja, pares $(x1, x2)$ cujo valor da f  prximo de 10.

Considere que "prximo" neste caso significa que o valor resultante da f no pode ser maior, em mdulo do que 1. Isto  $-1<f(x) - 10 < 1$, ou seja,  $9 <f(x)<11$

```python
def f(x1, x2):
    return x1**2 + x2**2

def create_meshgrid(x1_lims, x2_lims):
    x1mesh, x2mesh = np.meshgrid(x1_lims, x2_lims)
    return np.c_[x1mesh.ravel(), x2mesh.ravel()]

#Definindo tamanho dos vetores para o meshgrid

x1_lims = np.arange(-10,10,0.02)
x2_lims = np.arange(-10,10,0.02)

mesh = create_meshgrid(x1_lims, x2_lims)

L = []
W = []
x1_w = [] #para criar o circulo
x2_w = [] #para criar o circulo

for v in mesh:
    x1 = v[0]
    x2 = v[1]
    value_f = f(x1, x2)
    if 9 < value_f < 11:
        #w.append(value_f)
        x1_w.append(x1)
        x2_w.append(x2)
        
    L.append(value_f)
    
       
plt.figure(figsize=[15,10])
plt.scatter(mesh[:, 0], mesh[:, 1], c = L, cmap = plt.cm.rainbow) 
plt.scatter(x1_w,x2_w, c='white')
plt.colorbar()
plt.show()        
        
```
